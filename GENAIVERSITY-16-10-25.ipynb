```python
#4
import json

with open("C:/Users/prasa/Downloads/menu_items.json") as f:
    data = json.load(f)

multi_size_items = {}

for x in data:
    menu_items = x.get("menuItems", [])
    if not isinstance(menu_items, list):
        continue
    for y in menu_items:
        configs = y.get("customConfigs", [])
        if not isinstance(configs, list):
            continue
        sizes = {c.get("name") for c in configs if c.get("name")}
        if len(sizes) > 1:
            multi_size_items[y["name"]] = list(sizes)

print(multi_size_items)
```

```python
#5
import json

with open("C:/Users/prasa/Downloads/menu_items.json") as f:
    data = json.load(f)

wing_flavors = []

for x in data:
    for y in x.get("menuItems", []) or []:
        if y.get("name") == "Chicken Wings":
            for c in y.get("customConfigs", []) or []:
                for m in c.get("mandatoryModifiers", []) or []:
                    if m.get("name") == "Wing Flavor":
                        for mod in m.get("modifiers", []) or []:
                            wing_flavors.append(mod.get("name"))

print(wing_flavors)
```

```python
#8
import json

with open("C:/Users/prasa/Downloads/menu_items.json") as f:
    data = json.load(f)

max_price = 0
max_item = ""

for x in data:
    for y in x.get("menuItems", []) or []:
        for c in y.get("customConfigs", []) or []:
            price = c.get("itemPrice", 0)
            if price > max_price:
                max_price = price
                max_item = y.get("name", "Unknown")

print(f"Most Expensive Item: {max_item} - ${max_price / 100:.2f}")
```

```python
#9
import json

with open("C:/Users/prasa/Downloads/menu_items.json") as f:
    data = json.load(f)

price_ranges = {"Cheap": [], "Moderate": [], "Expensive": []}

for x in data:
    for y in x.get("menuItems", []) or []:
        for c in y.get("customConfigs", []) or []:
            price = c.get("itemPrice", 0) / 100  # convert cents to dollars
            name = y.get("name", "Unknown")
            if price < 8:
                price_ranges["Cheap"].append(name)
            elif 8 <= price <= 12:
                price_ranges["Moderate"].append(name)
            else:
                price_ranges["Expensive"].append(name)

# Remove duplicates
for k in price_ranges:
    price_ranges[k] = sorted(set(price_ranges[k]))

print("Cheap:", price_ranges["Cheap"])
print("Moderate:", price_ranges["Moderate"])
print("Expensive:", price_ranges["Expensive"])
```
